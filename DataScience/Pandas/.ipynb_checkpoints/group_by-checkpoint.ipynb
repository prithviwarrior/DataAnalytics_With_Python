```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("marks.csv")
```

```python
df
```

```python
def calc(record):
    return record.sum(), "Fail" if record.min()<35 else "Pass" 
```

```python
df[["Total","Result"]] = df[["M1", "M2", "M3"]].apply(calc, axis=1, result_type="expand")
```
