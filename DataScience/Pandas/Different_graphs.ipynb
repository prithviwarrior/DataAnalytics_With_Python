```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Comcast_telecom_complaints_data.csv")
```

```python
df
```

```python
results = df.groupby("Status")
```

```python
results.groups.keys()
```

```python
names=[]
case=[]
for Status, df2 in results:
    print("{}:{}".format(Status,len(df2)))
    names.append(Status)
    case.append(len(df2))
```

```python
highest_case_state = df.groupby("State")
```

```python
highest_case_state.groups.keys()
```

```python
l_states = []
l_cases = []
for Status, df3 in highest_case_state:
    print("{}:{}".format(Status,len(df3)))
    #df["Case_count"] = len(df) 
    l_states.append(Status)
    l_cases.append(len(df3))
    
```

```python
new_df = pd.DataFrame(
    {
        "State" : l_states,
        "Cases" : l_cases
    }
)
```
